```python
from statsmodels.graphics.tsaplots import plot_acf
import yfinance as yf
import pandas as pd 
import matplotlib.pyplot as plt 
from statsmodels.tsa.seasonal import seasonal_decompose
```

# Amazon Data

```python
amazon = yf.download('AMZN','2019-01-27', '2020-02-11')
amazon.head()
```

```python
plt.rcParams.update({'figure.figsize': (7,5)})
plt.plot(amazon["Open"])
```

```python
plot_acf(amazon['Open'], lags=20)

plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.plot(amazon["Open"])
plt.show()
```

```python
result = seasonal_decompose(amazon["Open"], model ="multiplicative", period=12)
```

```python
plt.rcParams.update({'figure.figsize': (8,10)})
result.plot()
plt.show()
```
